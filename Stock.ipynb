```python
import torch
import torch.nn as nn
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
```

```python
X = df1.iloc[:, :-3]
y = df1.iloc[:, -3]
X_train, X_test, y_train, y_test = train_test_split(X, y, train_size = 0.7, random_state = 777)
```

```python
X_train = torch.from_numpy(X_train.values).float()
y_train = torch.from_numpy(y_train.values).float().unsqueeze(dim=1)
X_test = torch.from_numpy(X_test.values).float()
y_test = torch.from_numpy(y_test.values).float().unsqueeze(dim=1)
```

## Try

```python
class Mod(nn.Module):
    def __init__(self, n):
        super(Mod, self).__init__()
        self.net = nn.Sequential(
                    nn.Linear(72, n),
                    nn.ReLU(),
                    nn.Linear(n, 1)
                    )
            
    def forward(self, x):
        return self.net(x)
```

```python
m1 = Mod(50).to(device)
optimizer = torch.optim.Adam(m1.parameters())
crit = nn.MSELoss()
```

```python
for epoch in range(1000):
    m1.train()
    for data in DataLoader(torch.cat((X_train, y_train), dim=1), batch_size = 100):
        optimizer.zero_grad()
        X_bp = data[:, :-1].to(device)
        y_bp = data[:, -1:].to(device)
        pre = m1(X_bp)
        loss = crit(pre, y_bp)
        loss.backward()
        optimizer.step()
```

```python
m1.eval()
total = 0
for data in torch.cat((X_test, y_test), dim=1):
    X_eva = data[:-1].to(device)
    y_eva = data[-1:].to(device)
    with torch.no_grad():
        pre = m1(X_eva)
        loss = crit(pre, y_eva)
    total += loss.cpu().item()
avg = total/len(X_test)
```

## Real

```python
lstm = nn.LSTM(1, 1, 4, batch_first=True).to(device)
optimizer = torch.optim.Adam(lstm.parameters())
crit = nn.MSELoss()
```

```python
for epoch in range(1000):
    lstm.train()
    for data in DataLoader(torch.cat((X_train, y_train), dim=1), batch_size = 100):
        optimizer.zero_grad()
        X_bp = data[:, :-1].unsqueeze(dim=2).to(device)
        y_bp = data[:, -1:].unsqueeze(dim=2).to(device)
        pre, (h, c) = lstm(X_bp)
        loss = crit(pre[:, -1:], y_bp)
        loss.backward()
        optimizer.step()
```

```python
lstm.eval()
total = 0
for data in torch.cat((X_train, y_train), dim=1):
    X_eva = data[:-1].unsqueeze(dim=1).to(device)
    y_eva = data[-1:].unsqueeze(dim=1).to(device)
    with torch.no_grad():
        pre, (h, c) = lstm(X_eva)
        loss = crit(pre[-1:], y_eva)
    total += loss.cpu().item()
avg = total/len(X_test)
```
